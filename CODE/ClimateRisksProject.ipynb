```python
def plot_column_frequency(data, column):
    fig, ax = plt.subplots(figsize=(16,8))
    frequency = data[column].value_counts()/len(data)
    frequency.plot(ax=ax, kind='bar')
```

```python
plot_column_frequency(data=geo_data, column='sec_type_1_name')
```

```python
plot_column_frequency(data=geo_data, column='asset_class2_name')
```

```python
plot_column_frequency(data=geo_data, column='geography_1_name')
```
